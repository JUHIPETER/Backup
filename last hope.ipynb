```python
import os
import sqlite3
from flask import jsonify
from flask_api import FlaskAPI
from flask_restful import Api
from sqlalchemy import create_engine
from flask import request
import json
app=FlaskAPI(__name__)
api=Api(app)
c=create_engine('sqlite:///PROJECT.db')

@app.route('/select', methods=['GET'])
def select():
    con = c.connect()
    y = con.execute("select * from sample")  
    return json.dumps([dict(r) for r in y])
    

@app.route('/insert', methods=['GET'])
def insert_tbl():
    con = c.connect()
    df.to_sql('sample', con,if_exists='replace')
    text = {"Message " : "Inserted Successfully"}
    return text
    
@app.route('/create', methods=['GET'])
def create():
    try:
        con = c.connect()
        trans = con.begin()
        con.execute('CREATE TABLE IF NOT EXISTS sample(SNO INT PRIMARY KEY,NAME TEXT,EMAIL TEXT,TEL TEXT,PRACTICE TEXT,POSTION TEXT,Timestamp DATETIME DEFAULT CURRENT_TIMESTAMP)')
        trans.commit()
        text = {"Message " : "Table Created successfully"}
        return text
    except Exception as error:
        text = {"Message " : "Table Creation Failed"}
        return text
    finally:
        con.close()
app.run()
```
